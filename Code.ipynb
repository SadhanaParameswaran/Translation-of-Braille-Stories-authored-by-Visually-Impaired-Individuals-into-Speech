```python
import os
from googletrans import Translator
import jiwer

folder_path = 'D:/College_Semesters/6th Semester/E. AI in Natural Language Processing/Project2/EnglishBooks'

original_stories = []

for file_name in os.listdir(folder_path):
    if file_name.endswith('.txt'):
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
            original_stories.append(text)

translator = Translator(service_urls=['translate.google.com'])

# Translate the stories to Tamil and Malayalam
translated_stories_tamil = []
translated_stories_malayalam = []

for story in original_stories:
    translated_text_tamil = translator.translate(story, dest='ta').text
    translated_text_malayalam = translator.translate(story, dest='ml').text
    translated_stories_tamil.append(translated_text_tamil)
    translated_stories_malayalam.append(translated_text_malayalam)

# Calculate WER for each story
for i in range(len(original_stories)):
    original_text = original_stories[i]
    translated_text_tamil = translated_stories_tamil[i]
    translated_text_malayalam = translated_stories_malayalam[i]
    wer_tamil = jiwer.wer(original_text, translated_text_tamil)
    wer_malayalam = jiwer.wer(original_text, translated_text_malayalam)
    print("Story", i+1)
    print("Tamil WER:", wer_tamil)
    print("Malayalam WER:", wer_malayalam)
    print()
```

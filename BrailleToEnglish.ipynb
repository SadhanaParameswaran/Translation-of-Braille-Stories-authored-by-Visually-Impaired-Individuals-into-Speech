```python
def braille_file_to_english(filename):
    braille_dict = {
        '⠁': 'a', '⠠⠁': 'A',
        
        '⠃': 'b', '⠠⠃': 'B',
        
        '⠉': 'c', '⠠⠉': 'C',
        
        '⠙': 'd', '⠠⠙': 'D',
        
        '⠑': 'e', '⠠⠑': 'E',
        
        '⠋': 'f', '⠠⠋': 'F',
        
        '⠛': 'g', '⠠⠛': 'G',
        
        '⠓': 'h', '⠠⠓': 'H',
        
        '⠊': 'i', '⠠⠊': 'I',
        
        '⠚': 'j', '⠠⠚': 'J',
        
        '⠅': 'k', '⠠⠅': 'K',
        
        '⠇': 'l', '⠠⠇': 'L',
        
        '⠍': 'm', '⠠⠍': 'M',  
        
        '⠝': 'n', '⠠⠝': 'N',
        
        '⠕': 'o', '⠠⠕': 'O',
        
        '⠏': 'p', '⠠⠏': 'P',
        
        '⠟': 'q', '⠠⠟': 'Q',
        
        '⠗': 'r', '⠠⠗': 'R',
        
        '⠎': 's', '⠠⠎': 'S',
        
        '⠞': 't', '⠠⠞': 'T',
        
        '⠥': 'u', '⠠⠥': 'U', 
        
        '⠧': 'v', '⠠⠧': 'V',
        
        '⠺': 'w', '⠠⠺': 'W',
        
        '⠭': 'x', '⠠⠭': 'X',
        
        '⠽': 'y', '⠠⠭': 'Y',
        
        '⠵': 'z', '⠠⠵': 'Z',

        '⠖': '!', 
        #'⠂': ',', 
        '⠲': '.', 
        '⠶': '(', 
        '⠶': ')', 
        '⠴': '”', 
        '⠦': '“', 
        '⠦': '?', 
        #'⠴': ' ',
        '⠤': '-', 
        '⠸⠣': '{',
        '⠸⠜': '}',
        '⠠⠶': '[',
        '⠶⠠': ']',
        '⠌': '/',
        '⠆': ';',
        '⠒': ':',

        '⠼⠁': '1', '⠼⠃': '2', '⠼⠉': '3', '⠼⠙': '4', '⠼⠑': '5', 
        '⠼⠋': '6', '⠼⠛': '7', '⠼⠓': '8', '⠼⠊': '9'#, '⠼⠚': '0'
    }
    with open(filename, 'r') as file:
        text = file.read()
    braille = ""
    for char in text:
        if char.isupper():
            braille += "⠠" + braille_dict[char.lower()]
        elif char in braille_dict:
            braille += braille_dict[char]
        else:
            braille += char
    return braille

braille = braille_file_to_english("D:\\College_Semesters\\6th Semester\\E. AI in Natural Language Processing\\Project2\\BrailleBooks\\MeetingAnimals_Braille.txt")

sourceFile = open('MyBestFriend_Braille_BrailleToAlpha.txt', 'w')
print(braille, file = sourceFile)
sourceFile.close()
```
